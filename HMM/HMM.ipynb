```python
from nltk.corpus import treebank
from nltk.tag import hmm
from nltk.tag import pos_tag
```

```python
# تحميل مجموعة بيانات treebank
nltk.download('treebank')
nltk.download('universal_tagset')
```

```python
# تحميل الجمل المصنفة في مجموعة البيانات
sentences = treebank.tagged_sents()
```

```python
# تقسيم البيانات إلى مجموعة تدريب واختبار
train_sents = sentences[:3000]
test_sents = sentences[3000:]
```

```python
# تدريب نموذج HMM
trainer = hmm.HiddenMarkovModelTrainer()
hmm_model = trainer.train(train_sents)
```

```python
# تطبيق النموذج بجملة جديدة لتصنيفها
test_sentence = "The quick brown fox jumps over the lazy dog".split()

# تصنيف الجملة
tagged_sentence = hmm_model.tag(test_sentence)
print(tagged_sentence)
```

```python
import numpy as np
from hmmlearn import hmm

# مجموعة بيانات التدريب (أمثلة)
# الكلمات والفئات المرتبطة بها
# كل عنصر في مجموعة التدريب هو (كلمة, فئة)
train_data = [
    ('في', 'PREP'),
    ('المؤتمر', 'NOUN'),
    ('الدولي', 'ADJ'),
    ('الأخير', 'ADJ'),
    ('حول', 'PREP'),
    ('الذكاء', 'NOUN'),
    ('الاصطناعي', 'ADJ'),
    ('قدم', 'VERB'),
    ('الدكتور', 'NOUN'),
    ('أحمد', 'NOUN'),
    ('السعيد', 'NOUN'),
    ('من', 'PREP'),
    ('جامعة', 'NOUN'),
    ('القاهرة', 'NOUN'),
    ('ورقة', 'NOUN'),
    ('بحثية', 'ADJ'),
    ('حول', 'PREP'),
    ('استخدام', 'NOUN'),
    ('الشبكات', 'NOUN'),
    ('العصبية', 'ADJ'),
    ('في', 'PREP'),
    ('تحليل', 'NOUN'),
    ('النصوص', 'NOUN'),
    ('باللغة', 'PREP'),
    ('العربية', 'ADJ'),
]

# إنشاء قائمة الكلمات والفئات للتدريب
words = [word for word, tag in train_data]
tags = [tag for word, tag in train_data]

# إنشاء قاموس للفئات مع أرقام
tagset = list(set(tags))
tag_to_id = {tag: i for i, tag in enumerate(tagset)}

# تحويل الفئات إلى أرقام
tags_numeric = [tag_to_id[tag] for tag in tags]
```

```python
#3. بناء وتدريب نموذج HMM
```

```python
# تعريف نموذج HMM
model = hmm.MultinomialHMM(n_components=len(tagset), random_state=42)

# إعداد بيانات التدريب
X = np.array([words.index(word) for word in words]).reshape(-1, 1)
lengths = [len(tags_numeric)]

# تدريب النموذج
model.fit(X, lengths)
```

```python
# توقع الفئات للنصوص الجديدة
```

```python
# نص جديد للتصنيف
new_text = ['الشبكات', 'العصبية', 'تستخدم', 'في', 'تحليل', 'النصوص']
X_new = np.array([words.index(word) if word in words else -1 for word in new_text]).reshape(-1, 1)

# توقع الفئات الجديدة
logprob, predicted_tags = model.decode(X_new, algorithm="viterbi")

# عرض النتائج
predicted_tags = [tagset[tag] for tag in predicted_tags]
for word, tag in zip(new_text, predicted_tags):
    print(f"{word}: {tag}")
```
