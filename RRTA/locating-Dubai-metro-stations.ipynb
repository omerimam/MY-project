### Using Folium to Visualize the Delhi Metro stations geographically

```python
#pip install folium
```

```python
import folium
```

```python
map1 = folium.Map(location=[latitude, longitude], zoom_start=11)

for lat, long, label in zip(df['latitude'], df['longitude'], df['station_name']):
    folium.CircleMarker(
        [lat,long],
        radius = 5,
        color = 'blue',
        popup = label,
        fill=True,
        fill_color='red',
        fill_opacity=0.6
    ).add_to(map1)
map1
```
